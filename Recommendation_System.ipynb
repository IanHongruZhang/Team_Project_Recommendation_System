```python
products_d = products.dropna(subset = ["title","description"])
```

```python
products_d.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
dtm = CountVectorizer(vocabulary = ["green","tea","curry","kit"]).fit_transform(products_d.title)
```

```python
x = dtm[0,:].toarray()
y = dtm[1,:].toarray()
```

```python
x = np.squeeze(x);y = np.squeeze(y)
x,y
```

```python
from math import sqrt
x.dot(y) / (sqrt(x.dot(x)) * sqrt(y.dot(y)))
x.dot(y) / sqrt((x.dot(x)) * y.dot(y))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(dtm)
```

```python
sim[0,:].argsort()[-10:]
```
