```python
import numpy as np
from numpy import NAN
#如果有缺失值，那该怎么办
#协同过滤
```

```python
m = np.matrix([[7,6,7,4,5,4],[6,7,NAN,4,3,4],[NAN,3,3,1,1,NAN],[1,2,2,3,3,4],[1,NAN,1,2,3,3]])
m
```

```python
mu = np.nanmean(m,axis = 1)
mu
```

```python
mc = m - mu
mc
```

```python
u = mc[0,:]
v = mc[2,:]
u = np.ravel(u)
v = np.ravel(v)
```

```python
Iuv = np.logical_and(np.isfinite(u), np.isfinite(v))
#取两个的交集
```

```python
#平方和，也就是每个的长度
norm = np.sqrt(np.sum(np.square(u[Iuv]))*np.sum(np.square(v[Iuv])))
```

```python
u[Iuv].dot(v[Iuv]) / norm
```

```python
def sparse_correlation(u,v):
    u = np.ravel(u)
    v = np.ravel(v)
    Iuv = np.logical_and(np.isfinite(u), np.isfinite(v))
    norm = np.sqrt(np.sum(np.square(u[Iuv]))*np.sum(np.square(v[Iuv])))
    return u[Iuv].dot(v[Iuv]) / norm
```

```python
sparse_correlation(u,v)
```

```python
#为这个矩阵的每行都计算它的相似值
sim = np.zeros(25).reshape(5,5)
for i in range(5):
    for j in range(5):
        sim[i,j] = sparse_correlation(mc[i,:],mc[j,:])
sim
```
